```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("loading The data under the name Trivago_data")
Trivago_data = pd.read_csv("hotel_bookings.csv")
```

```python
print(Trivago_data.head())
print("-"*100)
print(Trivago_data.columns)
```

```python
print(Trivago_data.shape)
print("-"*100)
print(Trivago_data.describe())
print(Trivago_data.info())
```

## EDA

```python
plt.figure(figsize=[35,12])
sns.catplot(x="is_canceled", y="lead_time", data=Trivago_data)
```

```python
fig = plt.figure(figsize = (10, 8))
sns.kdeplot(Trivago_data["lead_time"], shade=True)
plt.title("Density Plot of lead_time")
```

```python
sns.countplot(y="stays_in_weekend_nights", data=Trivago_data, color="c")
```

```python
sns.catplot(x="stays_in_weekend_nights", y="is_canceled", kind="bar", data=Trivago_data)
```

```python
fig, saxis = plt.subplots(2, 3,figsize=(16,12))
sns.barplot(x = 'adults', y = 'is_canceled', data=Trivago_data, ax = saxis[0,0])
sns.barplot(x = 'children', y = 'is_canceled', order=[1,2,3], data=Trivago_data, ax = saxis[0,1])
sns.barplot(x = 'babies', y = 'is_canceled', order=[1,0], data=Trivago_data, ax = saxis[0,2])
```

```python
plt.figure(figsize=[16,12])

plt.subplot(231)
plt.boxplot(x=Trivago_data['lead_time'], showmeans = True, meanline = True)
plt.title('lead time Boxplot')
plt.ylabel('lead time (days)')

plt.subplot(232)
plt.boxplot(Trivago_data['stays_in_weekend_nights'], showmeans = True, meanline = True)
plt.title('weekend nights Boxplot')
plt.ylabel('weekend nights (night)')

plt.subplot(233)
plt.boxplot(Trivago_data['stays_in_week_nights'], showmeans = True, meanline = True)
plt.title('week nights Boxplot')
plt.ylabel('week nights (night)')

plt.subplot(234)
plt.hist(x = [Trivago_data[Trivago_data['is_canceled']==1]['stays_in_weekend_nights'], Trivago_data[Trivago_data['is_canceled']==0]['stays_in_weekend_nights']], 
         stacked=True, color = ['c','#f08a5d'],label = ['canceled','not canceled'])
plt.title('stays_in_weekend_nights by is_canceled')
plt.xlabel('stays_in_weekend_nights (night)')
plt.ylabel('# ')
plt.legend()

plt.subplot(235)
plt.hist(x = [Trivago_data[Trivago_data['is_canceled']==1]['lead_time'], Trivago_data[Trivago_data['is_canceled']==0]['lead_time']], 
         stacked=True, color = ['c','#f08a5d'],label = ['canceled','not canceled '])
plt.title('lead_time Histogram by is_canceled')
plt.xlabel('lead_time (day)')
plt.ylabel('#')
plt.legend()

plt.subplot(236)
plt.hist(x = [Trivago_data[Trivago_data['is_canceled']==1]['stays_in_week_nights'], Trivago_data[Trivago_data['is_canceled']==0]['stays_in_week_nights']], 
         stacked=True, color = ['c','#f08a5d'],label = ['canceled','not canceled'])
plt.title('stays_in_week_nights Histogram by is_canceled')
plt.xlabel('stays_in_week_nights (#)')
plt.ylabel('#')
plt.legend()
```

```python
a = sns.FacetGrid(Trivago_data, hue = 'is_canceled', aspect=4 )
a.map(sns.kdeplot, 'stays_in_week_nights', shade= True )
a.set(xlim=(0 , Trivago_data['stays_in_week_nights'].max()))
a.add_legend()
```
